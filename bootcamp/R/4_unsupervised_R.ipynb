# Unsupervised Learning

```r
library(tidyverse)
library(dbscan)
library(factoextra)
```

## Titanic data: explore clusters



### Explore data + feature engineering

```r
titanic <- read.csv('../../data/titanic/titanic.csv')
head(titanic, 5)
```

```r
str(titanic)
```

```r
titanic_unsupervised <- titanic %>%
mutate(Cabin = as.numeric(Cabin),
      Embarked = as.numeric(Embarked),
      Sex = as.numeric(Sex),
      Survived = as.factor(Survived)) %>%
filter(is.na(Fare) == FALSE) 

titanic_unsupervised$Age[is.na(titanic_unsupervised$Age)] <- 
as.integer(mean(titanic_unsupervised$Age, na.rm = TRUE))

titanic_for_kmeans <- titanic_unsupervised%>%
select( -c(Name, PassengerId, Ticket, Survived))

head(titanic_for_kmeans, 3)
```

```r
summary(titanic_for_kmeans)
```

# kmeans clustering


### Build clusters + add to dataset

```r
k <- kmeans(titanic_for_kmeans, centers = 2, nstart = 10)

titanic_unsupervised$cluster <- k$cluster
head(titanic_unsupervised, 5)
```

### Evaluate clusters

```r
plot(titanic_for_kmeans, col = k$cluster)
```

```r
ggplot(titanic_unsupervised, aes(x = cluster, fill = Survived)) +
  #theme_bw() +
  geom_bar() +
  labs(x = "Cluster Assignment",
       y = "Passenger Count",
       title = "Titanic Training Survivability by Cluster")
```

# DBSCAN Clustering

Density-Based Spatial Clustering of Applications with Noise

### Build dbscan clusters

```r
db <- dbscan(titanic_for_kmeans, eps = .4, minPts = 25)
```

```r
db
```

```r
pairs(titanic_for_kmeans, col = db$cluster + 1L)
```

# Principal Component Analysis (PCA)


### Normalize Data

Rescale ("normalize") our earlier data so that it has a mean 0 and a standard deviation of 1

```r
titanic_for_pca <- scale(titanic_for_kmeans)

head(titanic_for_pca, 3)
```

```r
summary(titanic_for_pca)
```

```r
apply(titanic_for_pca, 2, sd)
```

### Fit PCA; Explore and Evaluate Results

```r
pca_fit <- prcomp(titanic_for_pca)

summary(pca_fit)
```

Along PC1, we're able to contain over 92% of variation.

```r
fviz_eig(pca_fit)
```

```r
fviz_pca_var(pca_fit,
             col.var = "contrib", # Color by contributions to the PC
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             )
```
